```python
a = open("US_Declaration.pdf", "rb")
pdf_reader = PyPDF2.PdfReader(a)
```

```python
first_page = pdf_reader.pages[0]  #page number index
```

```python
pdf_write = PyPDF2.PdfWriter()
```

```python
pdf_write.add_page(first_page)
```

```python
pdf_output = open("some_New_Doc.pdf","wb") #binary write  #ye naya name se pdf banaya ab isme add karenge pdf page jo write kiya hai 
```

```python
pdf_write.write(pdf_output)
```

```python
pdf_output.close()
a.close()
```

```python
f = open("US_Declaration.pdf","rb")  #read binary 
pdf_text = []  #list blank banaya 
pdf_reader = PyPDF2.PdfReader(f)

for p in range(len(pdf_reader.pages)):
    page = pdf_reader.pages[p]  #har ek page ko read karega phir add karega

    pdf_text.append(page.extract_text())

f.close()
```

```python
pdf_text
```

```python
for i in pdf_text:
    print(i)
    print("\n")
    print()
    print("\n")
    print()
    print("\n")
```
