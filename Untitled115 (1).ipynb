```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
Create a Doc object from the file peterrabbit.txt
```

```python
with open("peterrabbit.txt", "r") as  txt:
    text = txt.read()

#print(text)
```

```python
doc = nlp(text)
```

```python
print(doc)
```

```python
For every token in the third sentence, print the token text, the POS tag, the fine-grained TAG tag, and the description of the fine-grained tag
```

```python
#doc convert in list them make sentence 

sentence = len(list(doc.sents))
```

```python
sentence
```

```python
#Extract Sentence 3
sentence = list(doc.sents)[2]
print(sentence)
```

```python
for token in sentence:
    print(token.text, token.pos_, token.tag_, token.dep_)
```

```python
Provide a frequency list of POS tags from the entire document
```

```python
pos_freq = {} #create dcitonary

for token in doc:
    pos = token.pos_
    pos_freq[pos] = pos_freq.get(pos, 0) + 1

# Print POS tag frequencies
for pos, freq in pos_freq.items():
    print(f"{pos}: {freq}")
```
